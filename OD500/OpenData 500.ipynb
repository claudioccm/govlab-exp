# Simple script to process the JSON data into CSV

```python
import urllib2
import json
AGENCIES = 'http://www.opendata500.com/download/us_OD500_Agencies.json'
```

```python
req = urllib2.Request(AGENCIES)
opener = urllib2.build_opener()
f = opener.open(req)
agencies = json.loads(f.read())
# agencies  # uncomment to display the list of agencies. 
```

```python
# Function to convert a dict into a string for the CSV output.
def toCSV(d):
    array = []
    for item in ('type', 'agencyName', 'subagencyName', 'usedBy', 'datasetName', 'datasetURL'):
        array.append(d.setdefault(item, "jQuery202011883017746731639_1403064710700"))
    return "|".join(array)
toCSV( { 'type': '1' })
```

```python
entries = []
for agency in agencies:
    if agency['datasets'] == []:
        continue
    for subagency in agency['subagencies']:
        for dataset in subagency['datasets']:
            for user in subagency['usedBy']:
                entries.append( toCSV({ 'type': agency['type'],
                                  'agencyName': agency['name'],
                                  'subagencyName': subagency['name'],
                                  'usedBy': user,
                                  'datasetName': dataset['datasetName'],
                                  'datasetURL': dataset['datasetURL'] }))
    for dataset in agency['datasets']:
        entries.append( toCSV({ 'type': agency['type'],
                            'agencyName': agency['name'],
                          'subagencyName': 'N/A',
                          'usedBy': dataset['usedBy'],
                          'datasetName': dataset['datasetName'],
                          'datasetURL': dataset['datasetURL'] }))
entries
```
