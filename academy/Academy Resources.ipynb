This notebook lets you print the video content of the GovLab Academy wordpress site.

You can get the content by exporting the site as XML.

Then, you simply parse the XML and traverse it to find what you want.

Some attribute names are generated by the system and therefore have human unfriendly names; you need to create a mapping for that.

```python
import xml.etree.ElementTree as ET
import sys
import json

VIDEO_TITLE = '_simple_fields_fieldGroupID_20_fieldID_1_numInSet_0'
VIDEO_DESCRIPTION = '_simple_fields_fieldGroupID_20_fieldID_0_numInSet_0'

def processItem(item):
    title = item.find('title').text
    content_type = item.find('./wp:post_type', namespaces=namespaces).text
    url = None
    for post_meta in item.findall('.//wp:postmeta', namespaces=namespaces):
        meta_key = post_meta.find('./wp:meta_key', namespaces=namespaces)
        if meta_key.text == VIDEO_TITLE:
            url = post_meta.find('./wp:meta_value', namespaces=namespaces).text
        if meta_key.text == VIDEO_DESCRIPTION:
            description = post_meta.find('./wp:meta_value', namespaces=namespaces).text
    return { 'title': title, 'url': url, 'content_type': content_type, 'description': description }
```

```python
ET.register_namespace('wp', 'http://wordpress.org/export/1.2/')
namespaces = {'wp': 'http://wordpress.org/export/1.2/'}
# Make sure the notebook has permission to read this folder.
root = ET.parse("/Users/sahuguet/Downloads/thegovlabacademy.wordpress.2014-08-27.xml")
```

```python
all_video_items = root.findall(".//item[wp:post_type='video']", namespaces=namespaces)
videos = []
for item in all_video_items:
    videos.append(processItem(item))
```

```python
print "%d video(s) found." % len(videos)
print json.dumps(videos, indent=2)
```
